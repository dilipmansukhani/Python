head() and tail()

```python
#head and tail -->head upper k 5 row ka data nikal kar dedega
#   tial       --> tail last k 5 rows ka data nikal kar deta h

subs.head() # by default 5 rows dega
subs.head(3) # we can aslo adjust the no. of row
subs.head(7)
```
