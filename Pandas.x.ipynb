#Series from Lists
# It has 2 thing
  1.Index

  2.Values


```python
#String -- string se series bana skte h
country = ['india', 'pakistan','USA','Nepal','Srilanka']

pd.Series(country)   # S is caital on Series
```

```python
#integers -- integer se series bana sakte h
runs = [13,24,56,76,100]
pd.Series(runs)   # S is caital on Series
```

```python
#custom index --khud k hisab se index bana sakte ho
marks = [ 67,57,90,100]
subjects =["maths", "english","hindi", "science"]
pd.Series( marks, index=subjects )
```

```python
# setting a name --. to give name to your data
pd.Series( marks, index=subjects , name = "nitish k marks")
```

```python
#store -- now store this in
marks = pd.Series( marks, index=subjects , name = "nitish k marks")
marks   #print krne ki need ni h simply typ marks
```

#Series from dict.
sreies ko hum list se bhi sakte h

series ko dict. se bhi  bana sakte h




```python
marks = {
    "maths" :65,
    "english" : 67,
    "science" : 87,
    "hindi" :56
}

marks_series = pd.Series(marks,name= "nitish k marks")
marks_series
```

Series Attributes


```python
#Size --> tells the size of the series
marks_series.size
```

```python
#dtype  --> tells the  series dtype
marks_series.dtype
```

```python
#name  --> tells the series name
marks_series.name
```

```python
#is unique --> tells allseries item are unique or not
marks_series.is_unique  --> #writtern true

#if we create new series
pd.Series([1,1,1,2,3,5,5]).is_unique
```

```python
#index --> tels the all index
marks_series.index
```

```python
#value -->store value at index
marks_series.values
```
